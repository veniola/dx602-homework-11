### Problem 20

Set `p20` to be a data frame of data that groups the `titanic` data set on the age and class functions.
The average value for the remaining columns should be shown, and an additional column `size` should be added with the number of matching rows.
Then sort `p20` by the `size` column descending.


```python
# YOUR CHANGES HERE

# group by age and class, compute means for numeric columns only
numeric_cols = titanic.select_dtypes(include=[np.number]).columns.tolist()
# exclude grouping keys and target from aggregation
numeric_cols = [c for c in numeric_cols if c not in ("age", "class", "target")]
means = titanic.groupby(["age", "class"])[numeric_cols].mean().reset_index()
# compute group sizes
sizes = titanic.groupby(["age", "class"]).size().reset_index(name="size")
# merge means and sizes
p20 = means.merge(sizes, on=["age", "class"])
# sort by size column descending
p20 = p20.sort_values("size", ascending=False).reset_index(drop=True)
```
